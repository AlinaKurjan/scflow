```python
# read
adata = sc.read_h5ad(args['work_dir'] + 'scanpy_integ_1_qc.dir/adata_qc_unfilt.h5ad')
print(adata.obs['sample'].value_counts())
```

## Chromosome

```python
annot = sc.queries.biomart_annotations(
        "hsapiens",
        ["ensembl_gene_id", "external_gene_name", "start_position", "end_position", "chromosome_name"],
    ).set_index("external_gene_name")
annot
```

```python
chrY_genes =  pd.Index(adata.var['gene_name']).intersection(annot.index[annot.chromosome_name == "Y"])
chrY_gene_ids = adata.var.loc[adata.var['gene_name'].isin(chrY_genes)].index
chrY_gene_ids
# chrY percentage
adata.obs['pct_chrY'] = np.sum(adata[:, chrY_gene_ids].X, axis=1) / np.sum(adata.X, axis=1)
adata.obs
```

```python
# XIST counts
adata.obs['cnt_XIST'] = adata.X[:, adata.var['gene_name'].str.match('XIST')].toarray()
adata.obs
```

```python
sc.pl.scatter(adata, x='cnt_XIST', y='pct_chrY', color="sample", save='.pdf')
os.replace(os.path.join(os.getcwd(), 'figures/scatter.pdf'), os.path.join(os.getcwd(), 'figures/XIST_chrY_dot.pdf'))
shutil.move(os.path.join(os.getcwd(), 'figures/XIST_chrY_dot.pdf'), args['fig_work_dir'] + "XIST_chrY_dot.pdf")
shutil.rmtree(os.path.join(os.getcwd(), 'figures'))
```

```python
sc.pl.violin(adata, ["cnt_XIST", "pct_chrY"], jitter=0.4, groupby='sample', rotation=90, save='.pdf')
os.replace(os.path.join(os.getcwd(), 'figures/violin.pdf'), os.path.join(os.getcwd(), 'figures/XIST_chrY_dist.pdf'))
shutil.move(os.path.join(os.getcwd(), 'figures/XIST_chrY_dist.pdf'), args['fig_work_dir'] + "XIST_chrY_dist.pdf")
shutil.rmtree(os.path.join(os.getcwd(), 'figures'))
```
